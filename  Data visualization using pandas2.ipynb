```python
Townname=[]
Matches_played=[]
Towndata=df['home_team'].value_counts()
for _ in Towndata.index:
  Townname.append(_)
  Matches_played.append(Towndata[_])
```

```python
from matplotlib import pyplot as  plt
```

```python
plt.plot(Townname,Matches_played)
plt.xlabel('Town_name')
plt.ylabel('matchesplayed')
plt.title('matches played by town')
plt.show()
```

```python
plt.bar(Townname,Matches_played)
plt.ylabel('Matches_played')
plt.xlabel('Town Name')
plt.title('Matches Played by Towns')
plt.show()
```
